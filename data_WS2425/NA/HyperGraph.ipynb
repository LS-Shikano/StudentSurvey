```python
import pandas as pd
import hypernetx as hnx

df_w1 = pd.read_csv("../Cooked/df_w1_prepared.csv")
df_w2 = pd.read_csv("../Cooked/df_w2_prepared.csv")
df_w3 = pd.read_csv("../Cooked/df_w3_prepared.csv")
```

```python
id1 = set(df_w1['participant.label'])
id2 = set(df_w2['participant.label'])
id3 = set(df_w3['participant.label'])

complete = set(id1 & id2 & id3)
```

```python
df_socialmedia = df_w2.filter(["demographic_app.1.player.social_networks_1", 
                               "demographic_app.1.player.social_networks_2", 
                               "demographic_app.1.player.social_networks_3",
                               "demographic_app.1.player.social_networks_4",
                               "demographic_app.1.player.social_networks_5",
                               "demographic_app.1.player.social_networks_6",
                               "demographic_app.1.player.social_networks_7",
                               "demographic_app.1.player.social_networks_8",
                               "demographic_app.1.player.social_networks_9",
                               "demographic_app.1.player.social_networks_10",
                               "demographic_app.1.player.social_networks_11",
                               "demographic_app.1.player.social_networks_other",
                               ])
```

```python
hyper_dict = {}

for row in df_socialmedia.iterrows():
    for i in range(0, 4):
        if row[1][i] == 1: 
            hyper_dict.setdefault(i, []).append(row[0])
```

```python
H = hnx.Hypergraph(hyper_dict)

hnx.draw(H, with_edge_labels=False, with_node_labels=False)
```
