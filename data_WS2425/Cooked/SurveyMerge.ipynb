```python
import pandas as pd
```

```python
df_w1 = pd.read_csv("df_w1_prepared.csv")
df_w2 = pd.read_csv("df_w2_prepared.csv")

df_w1['R1'] = 1
df_w2['R2'] = 1
```

```python
sw1 = set(df_w1.columns)
sw2 = set(df_w2.columns)

sw1 ^ sw2
```

```python
id1 = set(df_w1['network_app.1.player.participantcode'])
id2 = set(df_w2['participant.label'])

len(id1 & id2)
```

```python
merged_df = pd.merge(df_w1, df_w2, left_on='network_app.1.player.participantcode', right_on='participant.label', how="inner", suffixes=("_df1", "_df2"))

# Loop through columns and combine columns with the same name into lists
for col in df_w1.columns:
    if col in df_w2.columns:  # Check if column name is the same in both DataFrames
        # Merge columns with the same name into a list
        merged_df[col] = merged_df.apply(
            lambda row: [x for x in (row.get(f'{col}_df1'), row.get(f'{col}_df2'))], axis=1
        )

        # Drop the intermediate columns (with suffixes)
        merged_df = merged_df.drop(columns=[f'{col}_df1', f'{col}_df2'])

merged_df
```
