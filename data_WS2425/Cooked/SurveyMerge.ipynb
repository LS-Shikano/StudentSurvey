```python
df_w1 = pd.read_csv("df_w1_prepared.csv")
df_w2 = pd.read_csv("df_w2_prepared.csv")
df_w3 = pd.read_csv("df_w3_prepared.csv")

df_w1['R1'] = 1
df_w2['R2'] = 1
df_w3['R3'] = 1
```

```python
df_bonus1 = df_w1[["participant.label", "R1", "end_app.1.player.rnumber", 'demographic_app.1.player.postcode']]
df_bonus2 = df_w2[["participant.label", "R2", "end_app.1.player.rnumber", 'demographic_app.1.player.postcode']]
df_bonus3 = df_w3[["participant.label", "R3", "end_app.1.player.rnumber", 'demographic_app.1.player.postcode']]
```

```python
merged_df1 = pd.merge(df_bonus1, df_bonus2, left_on='participant.label', right_on='participant.label', how="outer")
merged_complete = pd.merge(merged_df1, df_bonus3, left_on='participant.label', right_on='participant.label', how="outer")
```
